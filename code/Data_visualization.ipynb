```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
import pandas as pd
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
plt.tight_layout()
import matplotlib 
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 20}
matplotlib.rc('font', **font)
```

```python
df_remove_spare_columns = pd.read_csv("../data/data_remove_spare_columns.csv")
```

```python
num_instances = df_remove_spare_columns[df_remove_spare_columns.columns[0]].count()
```

```python
df_remove_spare_columns['Species'].unique()
```

```python
# Species percentage
species_list = list()
species_list.append((df_remove_spare_columns['Species'].values ==
                     'Adelie Penguin (Pygoscelis adeliae)').sum() / num_instances)
species_list.append((df_remove_spare_columns['Species'].values ==
                     'Chinstrap penguin (Pygoscelis antarctica)').sum() / num_instances)
species_list.append((df_remove_spare_columns['Species'].values ==
                     'Gentoo penguin (Pygoscelis papua)').sum() / num_instances)
plt.pie(species_list, labels = ['Adelie', 'Chinstrap', 'Gentoo'], 
        autopct='%1.1f%%', startangle=90)
```
